```python
# Imports
import pandas as pd
import numpy as np
from IPython.display import display

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
# Load data
ft_lem = pd.read_csv("./results/fasttext_lemmatized.csv", low_memory=False)
ft_swk = pd.read_csv("./results/fasttext_stopwords_kept.csv", low_memory=False)
ft_swr = pd.read_csv("./results/fasttext_stopwords_removed.csv", low_memory=False)

# Print length
print(len(ft_lem.index))
print(len(ft_swk.index))
print(len(ft_swr.index))

# Display
display(ft_lem)
display(ft_swk)
display(ft_swr)
```

```python
# Drop index column
ft_lem = ft_lem.drop(columns=['Unnamed: 0'])
ft_swk = ft_swk.drop(columns=['Unnamed: 0'])
ft_swr = ft_swr.drop(columns=['Unnamed: 0'])

# Extract just the numerical label from label and predicted_label column
ft_lem['label'] = [x.strip()[-1] for x in ft_lem['label']]
ft_lem['predicted_label'] = [x.strip()[-1] for x in ft_lem['predicted_label']]

ft_swk['label'] = [x.strip()[-1] for x in ft_swk['label']]
ft_swk['predicted_label'] = [x.strip()[-1] for x in ft_swk['predicted_label']]

ft_swr['label'] = [x.strip()[-1] for x in ft_swr['label']]
ft_swr['predicted_label'] = [x.strip()[-1] for x in ft_swr['predicted_label']]


# Convert to int type
ft_lem['label'] = ft_lem['label'].astype(int)
ft_lem['predicted_label'] = ft_lem['predicted_label'].astype(int)

ft_swk['label'] = ft_swk['label'].astype(int)
ft_swk['predicted_label'] = ft_swk['predicted_label'].astype(int)

ft_swr['label'] = ft_swr['label'].astype(int)
ft_swr['predicted_label'] = ft_swr['predicted_label'].astype(int)

# Display
display(ft_lem)
display(ft_swk)
display(ft_swr)
```

```python
# Compare accuracy of each clustering method
print("Classification metrics for each fastText dataset")
print("Accuracy score for ft_lem is:", 
      accuracy_score(ft_lem['label'], ft_lem['predicted_label']))

print("Accuracy score for ft_swk is:", 
      accuracy_score(ft_swk['label'], ft_swk['predicted_label']))

print("Accuracy score for ft_swr is:", 
      accuracy_score(ft_swr['label'], ft_swr['predicted_label']))

# Compare precision of each clustering method
print("\nPrecision score for ft_lem is:", 
      precision_score(ft_lem['label'], ft_lem['predicted_label']))

print("Precision score for ft_swk is:", 
      precision_score(ft_swk['label'], ft_swk['predicted_label']))

print("Precision score for ft_swr is:", 
      precision_score(ft_swr['label'], ft_swr['predicted_label']))

# Compare recall of each clustering method
print("\nRecall score for ft_lem is:", 
      recall_score(ft_lem['label'], ft_lem['predicted_label']))

print("Recall score for ft_swk is:", 
      recall_score(ft_swk['label'], ft_swk['predicted_label']))

print("Recall score for ft_swr is:", 
      recall_score(ft_swr['label'], ft_swr['predicted_label']))

# Compare f1 of each clustering method
print("\nF1 score for ft_lem is:", 
      f1_score(ft_lem['label'], ft_lem['predicted_label']))

print("F1 score for ft_swk is:", 
      f1_score(ft_swk['label'], ft_swk['predicted_label']))

print("F1 score for ft_swr is:", 
      f1_score(ft_swr['label'], ft_swr['predicted_label']))
```
